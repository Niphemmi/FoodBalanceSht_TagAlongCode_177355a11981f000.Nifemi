```python
#importing libraries
import pandas as np 
import pandas as pd 
```

```python
#importing datasets
food_balance_sht = pd.read_csv('FoodBalanceSheets_E_Africa_NOFLAG.csv', encoding = 'latin-1')
food_balance_sht.head()
```

```python
#checking for the no. of rows and columns in the df
food_balance_sht.shape
```

```python
#checking for each column info of the df
food_balance_sht.info()
```

```python
#checking for correlation between columns in the df
food_balance_sht.corr()
```

```python
#checking for the area with the highest sum in 2017
Y2017_group = food_balance_sht.groupby('Area')['Y2017'].sum()
sorted_2017 = Y2017_group.sort_values(ascending = False)
sorted_2017
```

```python
#checking for correlation between columns in the df
food_balance_sht.describe(include = 'all')
```

```python
#grouping by area for 2014
food_balance_sht.groupby('Area')['Y2014'].sum().sort_values(ascending = False)
```

```python
#getting mean and standard deviation across Y2017
mean_2017 = food_balance_sht['Y2017'].mean()
std_2017 = food_balance_sht['Y2017'].std()
round(mean_2017, 2)
round(std_2017, 2)
```

```python
#grouping by Element in each year to get the highest stock variation
food_balance_sht.groupby('Element')[['Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']].sum ()
```

```python
#checking for missing values in my df, focusing on year 2014
food_balance_sht.isna().sum()
missing_Y2014 = food_balance_sht['Y2014'].isna().sum()
total_count = food_balance_sht['Y2014'].count() + missing_Y2014

#calculating the percentage of the missing values
percentage_missing = (missing_Y2014 / total_count) * 100
round(percentage_missing, 3)
```

```python
#getting unique count of Areas
len(food_balance_sht['Area'].unique())
```

```python
#grouping by item for 2015 and 2018 to wine quantity
food_balance_sht.groupby('Item')[['Y2015', 'Y2018']].sum()
```

```python
#getting the 7th least sum among areas
Y2017_group = food_balance_sht.groupby('Area')['Y2017'].sum()
sorted_2017 = Y2017_group.sort_values(ascending = True)
sorted_2017.head(7)
```
